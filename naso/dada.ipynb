```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('search_result.csv', sep=';', encoding='ISO-8859-1')
data.head()

data_clear = data['institutions']

data_clear_2 = data['publicationDates']


data_clear.head()
data_clear

tab_data = [['UFRRJ', 187], ['UFAM',111],['PUC_RS',72],['UFSCAR',55],['UFBA',33]]
tab_tab_p = pd.DataFrame(tab_data, columns=['Universidade','Quantidade'])

tab_tab_p.head()
x = np.arange(len(tab_tab_p['Universidade']))

fig, ax = plt.subplots(figsize=(13,5))
bar_labels = tab_tab_p['Universidade']
bar_colors = ['#1976D2', '#C2185B', '#BBDEFB', '#F8BBD0', '#1EE76A']

ab = ax.bar(tab_tab_p['Universidade'],tab_tab_p['Quantidade'], label=bar_labels, color=bar_colors)

ax.set_ylabel('Quantidade')
ax.set_title('TOP 7 das Universidades no BDTD quando palavras chaves: "Necessidades", "Educativas" e "Especiais" ')
ax.legend(title='Universidade')
ax.set_xticks(x)

plt.show()

fig = plt.figure(figsize=(10,7))
ax.bar_label(ab, fmt="%.01f", size=10, label_type="edge")
plt.show()

data_clear_2.head()
plt.bar(data['publicationDates'],data['institutions'])
```
