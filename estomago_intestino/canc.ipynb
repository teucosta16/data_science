Leitura de dados

```python
#Leitura de dados homens

h_br = pd.read_csv('homens_brasil.csv',sep=';', encoding='ISO-8859-1')
#h_br.head()

h_c_o = pd.read_csv('homens_centro_oeste.csv', sep=';',encoding='ISO-8859-1')
#h_c_o.head()

h_nordeste = pd.read_csv('homens_nordeste.csv', sep=';',encoding='ISO-8859-1')
#h_nordeste.head()

h_norte = pd.read_csv('homens_norte.csv', sep=';',encoding='ISO-8859-1')
#h_norte.head()

h_sudeste = pd.read_csv('homens_sudeste.csv', sep=';',encoding='ISO-8859-1')
#h_sudeste.head()

h_sul = pd.read_csv('homens_sul.csv', sep=';',encoding='ISO-8859-1')
#h_sul.head()

#LEITURA DE DADOS MULHERES

m_br = pd.read_csv('mulheres_brasil.csv', sep=';', encoding='ISO-8859-1')
#m_br.head()

m_c_o = pd.read_csv('mulheres_centro_oeste.csv', sep=';', encoding='ISO-8859-1')
#m_br.head()

m_nordeste = pd.read_csv('mulheres_nordeste.csv', sep=';', encoding='ISO-8859-1')

m_norte = pd.read_csv('mulheres_norte.csv', sep=';', encoding='ISO-8859-1')

m_sudeste = pd.read_csv('mulheres_sudeste.csv', sep=';', encoding='ISO-8859-1')

m_sul = pd.read_csv('mulheres_sul.csv', sep=';', encoding='ISO-8859-1')

m_sul.head()

```

Tratamento dos dados e construção de tabela única - Homens
